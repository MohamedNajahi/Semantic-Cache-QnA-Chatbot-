```python
from fuzzywuzzy import fuzz

s1 = "apple iphone 15 pro"
s2 = "iphone 15 pro apple"

score = fuzz.ratio(s1, s2)

print(score)
```

```python
s1 = "how can i reset my password"
s2 = "how do i reset my password"

score = fuzz.ratio(s1, s2)

print(score)
```

```python
from fuzzywuzzy import fuzz

s1 = "apple iphone 15 pro"
s2 = "ipne pro apple 15"

score = fuzz.token_sort_ratio(s1, s2)

print(score)
```

```python
from fuzzywuzzy import process

query = "iphne 15"

choices = [
    "Samsung Galaxy S25",
    "Apple iPhone 15",
    "Google Pixel 10",
    "OnePlus 15"
]

best_match = process.extractOne(query, choices)

print(best_match)
```
